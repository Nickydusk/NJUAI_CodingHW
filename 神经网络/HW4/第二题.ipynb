```
# 载入数据集
import pandas as pd
import numpy as np

# 固定随机种子
def set_random_seed(seed):
    np.random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
set_random_seed(2020)

train = pd.read_csv('data/optdigits.tar',header=None).values
test = pd.read_csv('data/optdigits.tes',header=None).values
train_x = train[:,:64]
train_y = train[:,64]
test_x = test[:,:64]
test_y = test[:,64]
```

```
# 用BP神经网络进行分类
from sklearn.neural_network import MLPClassifier

for hidden_size in [1,5,10,20,50]:
  clf = MLPClassifier(solver = 'sgd',
            activation = 'relu',
            max_iter = 1000,
            hidden_layer_sizes = (hidden_size),
            random_state = 2020)
  clf.fit(train_x,train_y)
  print("hidden: %d  acc: %f"%(hidden_size,clf.score(test_x,test_y)))
```

```
clf = MLPClassifier(solver = 'sgd',
          activation = 'relu',
          max_iter = 1000,
          hidden_layer_sizes = (128,512,64),
          random_state = 2020)

from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
for train_indices, test_indices in kf.split(train_x):
  clf.fit(train_x[train_indices],train_y[train_indices])
  print(clf.score(train_x[test_indices],train_y[test_indices]))

print("hidden: 20,100,40 with 10_Fold_CrossValidation  acc: %f"%(clf.score(test_x,test_y)))
```
