```python
import numpy as np
import matplotlib.pyplot as plt

x = [0,20,40,60,80,100]
x = np.array(x)
print('x is :\n',x)
num = [33994,30077,26960,24643,23127,22410]
y = np.array(num)
print('y is :\n',y)

f1 = np.polyfit(x, y, 2)
print('f1 is :\n',f1)
p1 = np.poly1d(f1)
print('p1 is :\n',p1)
yvals = p1(x)
print('yvals is :\n',yvals)
```

```python
#绘图
plot1 = plt.plot(x, y, 's',label='original values')
#plot2 = plt.plot(x, yvals, 'r',label='polyfit values')
x_range = np.arange(0, 170, 5)
plot2 = plt.plot(x_range, p1(x_range), 'r', label='polyfit values')
plt.xlabel('offset')
plt.ylabel('score(mse)')
plt.legend(loc=1) 
plt.title('Polyfitting')
plt.savefig('Poly.png', dpi=300)
plt.show()
```
